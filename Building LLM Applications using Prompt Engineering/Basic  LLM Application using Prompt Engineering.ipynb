```python
def get_response(prompt,model="gpt-3.5-turbo"):
    messages=[{"role":"user","content":"prompt"}]
    response=openai.chat.completions.create(
        model=model,
        messages=messages
    )
    return response.choices[0].message["content"]
```

```python
prompt = input("")
print(get_response(prompt))
```

```python
history=[]
def chat(user_prompt,is_clear=False):
    global history
    if is_clear:
        history
    input_message={"role":"user","content":"user_prompt"}
    history.append(user_prompt)

    response=openai.chat.completions.create(
        model="gpt-3.5-turbo",
        messaqges= history
        )
    response =response.choices[0].messaqes["content"]
    response_messages={"role":"assistant","content":response}
    history.append(response_messages)
    return response
```
