```python
import yaml

with open('/content/gemini_key.yml','r') as file:
  api_creds = yaml.safe_load(file)
```

```python
import os

os.environ['GEMINI_API_KEY'] = api_creds['Gemini key']
```

```python
!pip install langchain-google-genai

from langchain_core.prompts import ChatPromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI

gemini_model = ChatGoogleGenerativeAI(model="gemini-pro-latest", convert_system_message_to_human=True)
```

```python
prompt = ChatPromptTemplate.from_template('{query}')
basic_chain = (
          prompt
          |
          gemini_model
)

response = basic_chain.invoke({"query": "What is the closest planet to the Sun"})

print(response.content)
```

```python
topics = [{'topic': 'AI'}, {'topic':'statistics'}]
responses = chain.map().invoke(topics)

for response in responses:
  print(response.content)
  print('----------------')
  print('/n')
```
